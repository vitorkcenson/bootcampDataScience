```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Inline indicates graphs should be presented as a cell output
%matplotlib inline
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
# Note the additional keep_default_na parameter which is set to False. This ensures
# that pandas does not interpret the None string in the AirBags column as NaN.
```

```python
car_data.head(10)
```

### 1. Boxplot

```python
# Boxplot for revs per mile
# A box plot for the revs per mile for the Audi, Hyundai, Suzuki, 
# and Toyota car manufacturers.

# Filter the DataFrame for the specified manufacturers
manufacturers = car_data[car_data['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])]

# Create the boxplot using seaborn
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=manufacturers)

# Customize the plot (optional)
plt.title('Boxplot of Revs per Mile for Selected Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Revs per Mile')

# Show the plot
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Based on the image, it appears that Toyota has the car with the highest revs per mile, as its boxplot extends the furthest to the right. Based on the image, it appears that Toyota has the car with the highest revs per mile, represented as a dot on this plot. 

### 2. Histogram

```python
# Histogram of MPG
# histogram of MPG in the city. On the same axis, show a histogram of
# MPG on the highway.

# Create the histograms
plt.hist(car_data['MPG.city'], alpha=0.8, label='City MPG')
plt.hist(car_data['MPG.highway'], alpha=0.6, label='Highway MPG')

# Add labels and title
plt.title("Histogram of MPG")
plt.xlabel('MPG')
plt.ylabel('Quantity')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
Based on the histogram, cars generally achieve better fuel efficiency (higher MPG) on the highway compared to city driving. This is likely due to factors like consistent speeds and reduced stop-and-go driving on highways.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

# Create a line plot with swapped axes
sns.lineplot(x='Turn.circle', y='Wheelbase', data=car_data)

# Add a title
plt.title('Relationship between Turning Circle and Wheelbase')
plt.xlabel('Turning Circle (ft)')  # Assuming units are in feet
plt.ylabel('Wheelbase (in)')  # Assuming units are in inches

# Show the plot
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
Positive Correlation: As the wheelbase increases, the turning circle tends to increase as well. This suggests a positive correlation between the two variables.

Variance: The shaded area around the line indicates the variability in turning circle for a given wheelbase. This suggests that other factors besides wheelbase also influence turning circle.

### 4. Barplot

```python
# Barplot for mean horsepower

# Create a bar plot
sns.barplot(x='Type', y='Horsepower', data=car_data)

# Add a title and labels
plt.title('Distribution of Horsepower by Car Type')
plt.xlabel('Car Type')
plt.ylabel('Horsepower')

# Show the plot
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
The bar chart shows the average horsepower for different car types. While larger cars like "Large" tend to have higher horsepower, there's no direct correlation. Midsize cars also have high horsepower, and smaller cars and vans generally have lower horsepower. This suggests that factors beyond size, such as car purpose and engine type, influence horsepower. Based on the bar chart, there is no clear and consistent relationship between car size and horsepower.
